**2. Data Quality Issues Handling**

1) Missing Values
   - Order Zipcode → 86.23% missing
   - Product Description → 100% missing
   - Customer Lname → 0.004% missing
   - Customer Zipcode → 0.0016% missing

```python
# Issue : 1 Completely Empty Columns
df.drop(columns=['Product Description'], inplace = True) # Remove it bcz only create noise

# Issue : 2 Highly missing values
df.drop(columns=['Order Zipcode'], inplace = True) #Too many missing values make it unreliable for modeling

# Issue : 3 Very low missing values
df['Customer Lname'] = df['Customer Lname'].fillna('Unknown')
df['Customer Zipcode'] = df['Customer Zipcode'].fillna(df['Customer Zipcode'].mode()[0]) # used mode bcz zipcode is categorical data

# Issue : 4 Redundant columns
df.drop(columns=['Customer Email','Customer Password','Customer Fname','Customer Street',
                 'Order Id', 'Order Item Id'], inplace=True) 
# These are identifiers or sensitive fields which do not contribute to predictive modeling

# Issue : 5 Incorrect Data Type
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])
# Required for : 
#   - Delivery time calculation
#   - Target variable creation
#   - Time-based feature engineerig 
```

**Q.2 Target Variable Creation**

**1.Binary Classification target for delivery performance**

---------------------------------------------------------------------------------
Columns used:
- Days for shipping (real)
- Days for shipment (scheduled)
---------------------------------------------------------------------------------
- if real shipping time > scheduled shipping time -> delivery is Late
- Else -> Delivery is On Time
---------------------------------------------------------------------------------
Target Variable Defination:
- Target = 1 -> Late Delivery
- Target = 0 -> On-Time Delivery

**2.Create target variable**

```python
df['target'] = (df['Days for shipping (real)'] > df['Days for shipment (scheduled)']).astype(int)
#used .astype(int) to convert bool dtype into int
```

**Q.3 Feature Engineering**

Engineer meaningful features that could help predict target variable, doen't include features directly reveal outcome.
- It may cause data leakage
- That means ML model will perform well on training data but fails in real world.

```python
# Feature 1: Order Month
# Seasonal demand may affect delivery performance
df['Order_Month'] = df['order date (DateOrders)'].dt.month

# Feature 2: Order day of week
# Weekend and weekdays may have different delivery patterns
df['Order_DayOfWeek'] = df['order date (DateOrders)'].dt.dayofweek

# Feature 3: Order Hour
# Order placed late at night may be processed later
df['Order_Hour'] = df['order date (DateOrders)'].dt.hour

# Feature 4: Discount Percentage
# Heavy discounts may increase order volume and cause delays
df['Discount_Percentage'] = df['Order Item Discount Rate'] * 100

# Feature 5: Total Order Quantity
# Larger order may take longer to process and ship
#df['Order Item Quantity']
```

**Q.4 Build ML Model and check predictions**

Step 1: Create x & y 

```python
# Remove the columns that cause data leakage(cause unrealistic high accuracy)
drop_cols = ['Days for shipping (real)','Days for shipment (scheduled)', 'Delivery Status', 
             'Late_delivery_risk','order date (DateOrders)', 'shipping date (DateOrders)']

# Create x (inputs [features/Questions]) & y (Output [target/Answers])
x = df.drop(columns=drop_cols + ['target'])
y = df.target
```

Step 2: Encode Categorical Columns

```python
# Dataset have categorical columns but Machine learning algorithms only understand numbers
# so we need to encode it, to make it model understandable

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in x.select_dtypes(include=['object','string']).columns:
    x[col] = le.fit_transform(x[col].astype(str))

# x.select_dtypes(include='object').columns -> Find all text columns
# le.fit_transform() -> convert text into numbers
# .astype(str) -> to avoid errors if column has mixed datatype
```

Step 3: Split the Dataset

-----------------------------------------------------
- Training Set: 80%
- Test Set: 20%

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42,
                                                    stratify = y)

# test_size=0.20: 20% data for testing, 80% data for training
# random_state=42: Fixed random split, everytime you run the code get the same results
# stratify=y: same proportion of train & test data -> without stratify we will get bad split like, 50-50, 90-10
#--------------keeps distribution balanced
```

Step 4: Train the Model

```python
# Algorithm : Random Forest
# Many decision trees vote together
# Final Answer : Majority Vote

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    random_state=42)

rf.fit(x_train, y_train)

# RandomForestClassifier() -> Creates the model
# .fit() -> teaches the model using training data
```

Step 5: Make Predictions 

```python
y_pred = rf.predict(x_test)  #predict classes 0 or 1
y_prob = rf.predict_proba(x_test)[:,1] #probability of being late used for ROC-AUC
```

Step 6: Calculate Metrics

| Metric    | Simple Meaning                                                  |
| --------- | --------------------------------------------------------------- |
| Accuracy  | Out of all orders, how many predictions were correct?           |
| Precision | Out of predicted late deliveries, how many were actually late?  |
| Recall    | Out of all actual late deliveries, how many did the model find? |
| F1-score  | Balance between Precision and Recall                            |
| ROC-AUC   | How well the model separates late and on-time deliveries        |


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)
print("ROC-AUC  :", roc_auc)
```

**Conclusion:**
The dataset was split into 80% training and 20% testing data using stratified sampling. A Random Forest Classifier was trained on the training set and evaluated on the test set using Accuracy, Precision, Recall, F1-score, and ROC-AUC metrics. The obtained scores indicate the model's ability to predict late deliveries.
